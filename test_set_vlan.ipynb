Менять имя порта


```python
from pysnmp.hlapi import *
from pysnmp.hlapi import *
from pysnmp.entity.rfc3413.oneliner import cmdgen
host = '10.1.124.104'
target = cmdgen.UdpTransportTarget((host, 161), timeout=2, retries=1)#куда подключаемся
snmp = cmdgen.CommandGenerator()
usm = cmdgen.UsmUserData('Admin', 'password01', 'password02')       # SNMP ver.3#аутентификация
errorIndication, errorStatus, errorIndex, varBinds = next(
    setCmd(SnmpEngine(),
           usm,
           target,
           ContextData(),
           ObjectType(ObjectIdentity('1.3.6.1.4.1.119.2.3.69.501.5.30.1.1.11.8650752'),
                      OctetString('test')))
)
#ObjectType(ObjectIdentity('1.3.6.1.4.1.119.2.3.69.501.5.30.1.1.11.8650752' имя порта
#ObjectType(ObjectIdentity('1.3.6.1.4.1.119.2.3.69.5.3.1.1.3.1') имя IDU
#OctetString('test') Вводим имя порта/idu, все знечения которые строка
#Блок обработки ошибок
if errorIndication:
    print(errorIndication)
elif errorStatus:
    print('%s at %s' % (errorStatus.prettyPrint(),
                        errorIndex and varBinds[int(errorIndex) - 1][0] or '?'))
else:
    for varBind in varBinds:
        print(host,' = '.join([x.prettyPrint() for x in varBind]))
```

Создание VLAN

```python
import pysnmp
from pysnmp.hlapi import *
from pysnmp.entity.rfc3413.oneliner import cmdgen


host = '10.1.115.21'
oidVlan = ('1.3.6.1.4.1.119.2.3.69.501.5.20.5.1.4.3010')
print(oidVlan)
oidVlanName = '1.3.6.1.4.1.119.2.3.69.501.5.20.5.1.3.3010' 
print(oidVlanName)
vlan = int(123)
snmp = cmdgen.CommandGenerator()
target = cmdgen.UdpTransportTarget((host, 161), timeout=2, retries=1)#куда подключаемся
usm = cmdgen.UsmUserData('Admin', 'password01', 'password02')   
errorIndication, errorStatus, errorIndex, varBinds = next(
    setCmd(
        SnmpEngine(),
        usm,
        target,
        ContextData(),
        ObjectType(ObjectIdentity(oidVlan), Integer(4)),#4-запись; 6-удаление; 1 - просто статус внутрений.
        ObjectType(ObjectIdentity(oidVlanName), OctetString("LTE-4647_F18(m)"))))

if errorIndication:
    print(errorIndication)
elif errorStatus:
    print('%s at %s' % (errorStatus.prettyPrint(),
                        errorIndex and varBinds[int(errorIndex) - 1][0] or '?'))
else:
    for varBind in varBinds:
        print(host,' = '.join([x.prettyPrint() for x in varBind]))
```

```python
import pysnmp
from pysnmp.hlapi import *
from pysnmp.entity.rfc3413.oneliner import cmdgen


host = '10.1.124.104'
oidVlan = ('1.3.6.1.4.1.119.2.3.69.501.5.20.5.1.4.12')
print(oidVlan)
oidVlanName = '1.3.6.1.4.1.119.2.3.69.501.5.20.5.1.3.12' 
print(oidVlanName)
vlan = int(123)
snmp = cmdgen.CommandGenerator()
target = cmdgen.UdpTransportTarget((host, 161), timeout=2, retries=1)#куда подключаемся
usm = cmdgen.UsmUserData('Admin', 'password01', 'password02')    

for (errorIndication, errorStatus,errorIndex, varBinds) in bulkCmd(
    SnmpEngine(),usm,target,ContextData(),0, 50,  # GETBULK specific: request up to 50 OIDs in a single response
    ObjectType(ObjectIdentity('1.3.6.1.4.1.119.2.3.69.501.5.20.5.1'))):
    if errorIndication:
        print(errorIndication)
        break
    elif errorStatus:
        print('%s at %s' % (errorStatus.prettyPrint(),
            errorIndex and varBinds[int(errorIndex)-1][0] or '?'))
        break
    else:
        for varBind in varBinds:
            print(' = '.join([x.prettyPrint() for x in varBind]))       
 
```
