Задание 2.1¶ 
1) Сгенерирйуте данные: рассчитайте значения функций $y = x^3$ и $y = 2*x^3$ на отрезке от -10 до 10 с шагом 1.

2)Постройте графики функции на одном полотне и добавьте легену, указав формулой вид построенных функций. Залейте область между двумя функциями цветом, используя при заливке интенсивность цвета 0.5. Убедитесь, что по оси X значения изменяются в диапазоне [-10, 10].


```python
import numpy as np
import matplotlib.pyplot as plt
#matplotlib.pyplot отслеживает состояние текущего изображения, и области построения, 
##так что отрисовывающие функции применяются к текущим осям.
x=np.arange(-10,11,1)
test=x**3
test2 = 2*(x**3)
plt.plot(x,test,label='$y=x^3$',linewidth=2.0)
plt.plot(x,test2,label='$y=2*x^3$',linewidth=2.0)
plt.fill_between(x, test, test2, color='blue', alpha=0.5, label='place between')
plt.legend(loc = 9)
plt.show()
```

Задание 2.2 
1) Сгенерирйуте данные: рассчитайте значения функций $y = x^2$ и $y = 2*x^2$ на отрезке от -10 до 10 с шагом 1.

2) Сгенерирйуте данные: рассчитайте значения функций $y = x^3$ и $y = 2*x^3$ на отрезке от -10 до 10 с шагом 1.

3) Постройте изображение, состоящее из двух изображений, расположенных рядом горизонтально (используйте plt.subplot). На первом изображении постройте графики функции из пункта 1) и добавьте легену, указав формулой вид построенных функций. Залейте область между двумя функциями цветом, используя при заливке интенсивность цвета 0.5. На втором изображении постройте графики функции из пункта 2) и добавьте легену, указав формулой вид построенных функций. Залейте область между двумя функциями цветом, используя при заливке интенсивность цвета 0.5.


```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(-10,11,1)
f1_2=x*x
f2_2 = 2*(x**2)
f1_3=x**3
f2_3 = 2*(x**3)
plt.figure(figsize=(6,8))

plt.subplot(211)
plt.plot(x,f1_2,label='$y=x^2$',linewidth=2.0 )
plt.plot(x,f2_2,label='$y=2*x^2$',linewidth=2.0 )
plt.fill_between(x, f1_2 , f2_2, color='blue', alpha=0.5, label='place between')
plt.title('$y=x^2$'+' and '+'$y=2*x^2$' )
plt.legend(loc = 9)

plt.subplot(212)
plt.plot(x,f1_3,label='$y=x^3$',linewidth=2.0 )
plt.plot(x,f2_3,label='$y=2*x^3$',linewidth=2.0 )
plt.fill_between(x, f1_3 , f2_3, color='red', alpha=0.5, label='place between')
plt.title('$y=x^3$'+' and '+'$y=2*x^3$')
plt.legend(loc = 9)

plt.show()
```

Задание 3.1 
1) Загрузите набор данных "Breast Cancer Wisconsin (Diagnostic) Database" с помщью модуля datasets из библиотеки sklearn. Изпользуйте метод load_breast_cancer()

2) Выведите на экран и ознокомьтесь с описанием этого набора данных


```python
import pandas as pd
from sklearn import datasets
data=datasets.load_breast_cancer()##загрузка
print data.DESCR ##Выведите на экран и ознокомьтесь с описанием этого набора данных
```

```python
data.data
```

```python
data.target_names
```

```python
data_frame = pd.DataFrame(data.data)
data_frame.columns = data.feature_names
data_frame['target'] = data.target
data_frame.target = data_frame.target
```

```python
data_frame.head()
```

```python
plt.pie(data_frame.target.value_counts(),
        explode=[0, 0.2], autopct='%1.1f%%', labels=data['target_names'])
plt.show()
```

```python
from matplotlib.colors import ListedColormap
colors = ListedColormap(['red', 'blue'])
plt.figure(figsize=(12,4))

plt.subplot(121)
plt.scatter(data_frame.loc[:,'mean radius'], data_frame.loc[:,'mean texture'],
            c = data_frame.target,cmap = colors)
plt.xlabel('mean radius')
plt.ylabel('mean texture')

plt.subplot(122)
plt.scatter(data_frame.loc[:,'mean perimeter'],data_frame.loc[:,'mean area'], 
            c = data_frame.target, cmap = colors)
plt.xlabel('mean perimeter')
plt.ylabel('mean area')
plt.show()
```
