```python
# coding: utf-8
import time
import telnetlib
text = str()
user = "admin"
password = "admin"
a="Previously saved configuration archive removed successfully"
with open('../Reboot/ip.txt', 'r') as ff:
    a = ff.readlines()
    a = [line.rstrip() for line in a]#удаляем\n
    a = ",".join(a)
    for host in a.split(','):
        # Логинимся
        tn = telnetlib.Telnet(host)
        print 'login',
        tn.read_until("Username: ")
        tn.write(user + "\r")
        print 'ok\npass',
        tn.read_until("Password: ")
        tn.write(password + "\r")
        print 'ok'
        
        # Читаем результат, посылаем команду
        
        time.sleep(3)
        
        print tn.read_some()
        
        #Сохраняем конфигурацию железки
        start='start'+' '+time.asctime()
        print(host+"_"+start)
        tn.write("config save" + "\r") #сохранение всего
        tn.read_until("Previously saved configuration archive removed successfully")
        rez=host+'_'+start+'_'+'_'+'config save: '+'_'+'ok'+'_'+time.asctime(),'\n'
        print(time.asctime())
        for d in rez:
            text+=d
        
        #Рестар NE
        start='start'+' '+time.asctime()
        print(host+"_"+start+'reboot')
        tn.write("reboot system" + "\r") #reboot
        tn.read_until("Are you sure you want to continue? (y/n) ")
        tn.write("y" + "\r")
        rez=host+'_'+start+'_'+'_'+'reboot '+'_'+'ok'+'_'+time.asctime(),'\n'
        for d in rez:
            text+=d
            
        #Задержка проверка, что NE доступен
        time.sleep(600)
        
        tn = telnetlib.Telnet(host)
        print 'login',
        tn.read_until("Username: ")
        tn.write(user + "\r")
        print 'ok\npass',
        tn.read_until("Password: ")
        tn.write(password + "\r")
        print 'ok'
        
        # Читаем результат, посылаем команду
                    
        print tn.read_some()
        
        start='start'+' '+time.asctime()
        print(host+"_"+start)
        tn.write("config save" + "\r") #сохранение всего
        tn.read_until("Previously saved configuration archive removed successfully")
        rez=host+'_'+start+'_'+'_'+'config save: '+'_'+'ok'+'_'+time.asctime(),'\n'
        print(time.asctime())
        for d in rez:
            text+=d
              
with open('../Reboot/rezultat.csv', 'w') as f:
    f.write(text)
print unicode ('Выполнил',encoding='utf-8')
```
